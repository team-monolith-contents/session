# Ipydoodle을 활용한 시각화 수업

## World()

**Ipydoodle**을 시작하기 위해서는 가장 먼저 **World**를 만들어 주어야 합니다.

**Ipydoodle**의 모든 물체들은 **World** 안에 존재합니다.

**World**는 `World()` 명령어를 통해 만들 수 있습니다.

`World()` 명령어를 통해서 가장 간단한 기본 상태(default)의 **World**를 만들 수 있습니다.

기본 상태의 **World**는 흰색의 가로(width) 700, 세로(height) 500의 크기를 가집니다. 또한 축(axis)도 가지고 있습니다.

World의 크기나 색, 축은 World의 속성을 지정해 주는 것으로 바꿀 수 있습니다.

### World
- **width** : 가로 길이
- **height** : 세로 길이
- **color** : Box의 색
- **axis** : 축의 유무(True : 축이 있음, False : 축이 없음)
- **axis_color** : 축의 색

**Ipydoodle**에서는 다양한 색을 사용할 수 있습니다.



```python
# 자신만의 World()를 만들고 속성값들을 바꿔보세요!

World(height = 200, width = 500, color = '노란색', axis = True, axis_color = 'red') # World 를 만들어 줍니다.
```

## Object

**Ipydoodle**에서는 **Box**, **Circle**, **Line**의 3가지 Object를 만들 수 있습니다.

각각의 object는 World와 마찬가지로 간단한 명령어 `Box()`, `Circle()`, `Line()`을 통해 만들 수 있습니다.

또한 각각의 위치나 모양, 색을 속성값을 바꾸어주어 변경할 수 있습니다.

각각의 물체들은 아래와 같은 속성을 가집니다.

### Box
- **width** : 가로 길이
- **height** : 세로 길이
- **x** : 중심의 X 좌표
- **y** : 중심의 Y 좌표
- **color** : Box의 색
- **alpha** : Box의 투명도(0~1사이의 실수 값, 1에 가까울 수록 불투명, 0에 가까울 수록 투명)

### Circle
- **radius** : 반지름
- **x** : 중심의 X 좌표
- **y** : 중심의 Y 좌표
- **color** : Circle의 색
- **alpha** : Circle의 투명도(0~1사이의 실수 값, 1에 가까울 수록 불투명, 0에 가까울 수록 투명)

### Line
- **x1** : 첫 번째 점의 X 좌표
- **y1** : 첫 번째 점의 Y 좌표
- **x2** : 두 번째 점의 X 좌표
- **y2** : 두 번째 점의 Y 좌표
- **color** : Line의 색
- **alpha** : Line의 투명도(0~1사이의 실수 값, 1에 가까울 수록 불투명, 0에 가까울 수록 투명)

```python
# 자신만의 방법으로 여러 물체들을 만들어봅시다.

World()
Box(width = 200, height = 400, x= 100 , y = 100, color = 'cyan', alpha = 0.5) # Box를 만듭니다.
Circle(x = 100, y = 100, radius = 10, color = 'red', alpha = 0.8) # Circle을 만듭니다.
Line(x1 = 100, y1 = 100, x2 = 200, y2= 200, color = 'blue', alpha = 0.3) # Line을 만듭니다.
```

## 체크무늬 만들기 실습

이제 World와 물체를 만드는 방법에대해 배웠으니 이를 활용해 아래의 그림과 같은 체크무늬를 만드는 실습을 해보도록 하겠습니다.

체크무늬는 기초파이썬 시간에 배웠던 반복문과 조건문을 통해 만들 수 있습니다.



체크무늬만들기는 다음과 같은 순서로 진행됩니다.

1. 회색 배경에 box(50 x 50)를 만듭니다. 
2. 7개의 박스를 가로로 만듭니다.
3. 7개의 박스의 색이 번걸아 가면서 나오도록 합니다.
4. 가로 세로 7개의 박스가 나오도록 합니다.(가로로 색이 번갈아 나오는 것을 유지합니다.)
5. 가로 세로 7개의 박스가 나오도록 합니다. 세로로도 색이 번갈아 나오도록 합니다.

```python
# 회색 배경에 box(50 x 50)를 만듭니다. 

World(color = 'gray')

Box(width = 50, height = 50)
```

```python
# 7개의 박스를 가로로 만듭니다.

World(color = 'gray')

for i in range(7):
    Box(x = 50*i-150, width = 50, height = 50)
```

```python
# 7개의 박스의 색이 번걸아 가면서 나오도록 합니다.

World(color = 'gray')

even_color = 'black'
odd_color = 'white'

for i in range(7):
    if i % 2 ==0:
        Box(x = 50*i-150, width = 50, height = 50, color = even_color)
    else:
        Box(x = 50*i-150, width = 50, height = 50, color = odd_color)
```

```python
# 가로 세로 7개의 박스가 나오도록 합니다.(가로로 색이 번갈아 나오는 것을 유지합니다.)

World(color = 'gray')

even_color = 'black'
odd_color = 'white'

for i in range(7):
    for j in range(7):
        if i % 2 ==0:
            Box(x = 50*i-150, y = 50*j-150, width = 50, height = 50, color = even_color)
        else:
            Box(x = 50*i-150, y = 50*j-150, width = 50, height = 50, color = odd_color)
```

```python
# 가로 세로 7개의 박스가 나오도록 합니다. 세로로도 색이 번갈아 나오도록 합니다.

World(color = 'gray')

for i in range(7):
    for j in range(7):
        if j % 2 == 0:        
            even_color = 'black'
            odd_color = 'white'
        else:
            even_color = 'white'
            odd_color = 'black'            
        if i % 2 == 0 :
            Box(x = 50*i-150, y = 50*j-150, width = 50, height = 50, color = even_color)
        else:
            Box(x = 50*i-150, y = 50*j-150, width = 50, height = 50, color = odd_color)
```

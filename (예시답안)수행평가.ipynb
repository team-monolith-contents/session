## World()

**Ipydoodle**을 시작하기 위해서는 가장 먼저 **World**를 만들어 주어야 합니다.

**Ipydoodle**의 모든 물체들은 **World** 안에 존재합니다.

**World**는 `World()` 명령어를 통해 만들 수 있습니다.

`World()` 명령어를 통해서 가장 간단한 기본 상태(default)의 **World**를 만들 수 있습니다.

기본 상태의 **World**는 흰색의 가로(width) 700, 세로(height) 500의 크기를 가집니다. 또한 축(axis)도 가지고 있습니다.

World의 크기나 색, 축은 World의 속성을 지정해 주는 것으로 바꿀 수 있습니다.

### World
- **width** : 가로 길이
- **height** : 세로 길이
- **color** : Box의 색
- **axis** : 축의 유무(True : 축이 있음, False : 축이 없음)
- **axis_color** : 축의 색

**Ipydoodle**에서는 다양한 색을 사용할 수 있습니다.



## Object

**Ipydoodle**에서는 **Box**, **Circle**, **Line**의 3가지 Object를 만들 수 있습니다.

각각의 object는 World와 마찬가지로 간단한 명령어 `Box()`, `Circle()`, `Line()`을 통해 만들 수 있습니다.

또한 각각의 위치나 모양, 색을 속성값을 바꾸어주어 변경할 수 있습니다.

각각의 물체들은 아래와 같은 속성을 가집니다.

### Box
- **width** : 가로 길이
- **height** : 세로 길이
- **x** : 중심의 X 좌표
- **y** : 중심의 Y 좌표
- **color** : Box의 색
- **alpha** : Box의 투명도(0~1사이의 실수 값, 1에 가까울 수록 불투명, 0에 가까울 수록 투명)

### Circle
- **radius** : 반지름
- **x** : 중심의 X 좌표
- **y** : 중심의 Y 좌표
- **color** : Circle의 색
- **alpha** : Circle의 투명도(0~1사이의 실수 값, 1에 가까울 수록 불투명, 0에 가까울 수록 투명)

### Line
- **x1** : 첫 번째 점의 X 좌표
- **y1** : 첫 번째 점의 Y 좌표
- **x2** : 두 번째 점의 X 좌표
- **y2** : 두 번째 점의 Y 좌표
- **color** : Line의 색
- **alpha** : Line의 투명도(0~1사이의 실수 값, 1에 가까울 수록 불투명, 0에 가까울 수록 투명)
